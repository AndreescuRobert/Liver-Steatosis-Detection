```python
import scipy.io
import pandas as pd
from cv2 import imshow
import matplotlib.pyplot as plt
from PIL import Image

mat = scipy.io.loadmat('dataset_liver_bmodes_steatosis_assessment_IJCARS.mat')


print(mat.keys())

data = [[row.flat[0] for row in line] for line in mat['data'][0]]

data2 = []
data2_main = []
place = 0
row_num = 0
image_class = 0
for line in mat['data'][0]:
    for row in line:
        place += 1
        if place % 2 == 0:
            image_class = row.flat[0]
        if place % 4 == 0:
            row_num += 1
            for image_i in range(10):
                # print(image_i)
                image = row[image_i]
                # plt.imshow(image)
                # plt.savefig(string)
                image_class = 1 if row_num > 17 else 0
                string = "datasets/orig/" + "10000" + "/" + str(image_class) + "/" + str(row_num) + "_" + str(image_i) + "_class" + str(image_class) + ".png"

                data = Image.fromarray(image)

                # saving the final output
                # as a PNG file
                data.save(string)
                # print("&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&")
        data2.append(row.flat[0])
    data2_main.append(data2)
    data2 = []
columns = ['id', 'class', 'fat', 'images']
df_train = pd.DataFrame(data2_main, columns=columns)
print(df_train)
```
